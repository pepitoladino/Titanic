```python
# Importar  librerias
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Leer y crear el DataFrame
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
print(df)
```

```python
# Imprimir la informacion
print(df.info())
```

```python
# Analisis de valores faltantes
plt.figure(figsize=(10,8))
sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
# Vizualisando los datos
plt.figure(figsize=(10,8))
#sns.heatmap(df.corr(), annot=False, cmap="coolwarm")
```

```python
# Distribucion de la edad
plt.figure(figsize=(10,8))
sns.histplot(df['Age'].dropna(), kde=False, bins=30)
```

```python
# sobrevivientes por genero
plt.figure(figsize=(10,8))
sns.countplot(x='Survived', hue='Sex', data=df)
```

```python
# Sobrevivientes por clase
plt.figure(figsize=(10, 8))
sns.countplot(x='Survived', hue='Pclass', data=df)
```

```python
# Ditribucion de la tarifa
plt.figure(figsize=(10, 8))
sns.histplot(df['Fare'], kde=False, bins=40)
```

```python
# Relacion entre la edad y la clase
plt.figure(figsize=(10, 8))
sns.boxplot(x='Pclass', y='Age', data=df)
```

```python
# Relacion entre la edad y genero
plt.figure(figsize=(10, 8))
sns.boxplot(x='Sex', y='Age', data=df)
```

```python
# Relacion entre la tarifa y la clase
plt.figure(figsize=(10, 8))
sns.boxplot(x='Pclass', y='Fare', data=df)
```

```python
# Creando una nueva caracteristca 'Familysize' que combine 'SibSp' y 'Parch'
df['Familysize'] = df['SibSp'] + df['Parch'] + 1
```

```python
# Creando una nueva caracteristica 'IsAlone' que indica si el psajero est solo
df['IsAlone'] = 0
df.loc[df['Familysize'] == 1, 'IsAlone'] = 1
```

```python
# Creando una nueva caracteristica 'AgeClass' que combine 'Age' y 'Pclass'
df['AgeClass'] = df['Age'] * df['Pclass']
```

```python
# Creando una nueva caracteritica 'FarePerPerson' que divide 'Fare' y 'Familysize'
df['FarePerPerson'] = df['Fare'] / df['Familysize']
```

```python
# Visualizando las nuevas caracteriticas
plt.figure(figsize=(10, 8))
#sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
```

```python
plt.figure(figsize=(10, 8))
sns.countplot(x='IsAlone', hue='Survived', data=df)
```

```python
plt.figure(figsize=(10, 8))
sns.histplot(df['Familysize'], kde=False, bins=30)
```

```python
plt.figure(figsize=(10, 8))
sns.histplot(df['AgeClass'].dropna(), kde=False, bins=30)
```

```python
plt.figure(figsize=(10, 8))
sns.histplot(df['FarePerPerson'], kde=False, bins=30)
```
